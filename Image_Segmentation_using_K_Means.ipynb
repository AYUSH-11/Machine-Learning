```python
def calc_distance(x,y):         #calculate distance between two points
    return (sum((x-y)**2))**0.5
```

```python
def findclosetcentroids(c,x):
    assign_centroid=[]
    for i in x:
        distance=[]
        for j in c:
            distance.append(calc_distance(i,j))
        assign_centroid.append(np.argmin(distance))        # np.argmin return index of min element
    return assign_centroid
```

```python
def calc_centroids(cluster,x):
    new_df1=pd.concat([pd.DataFrame(x),pd.DataFrame(cluster,columns=['cluster'])],axis=1) # Add one column which tells cluter of point
    new_centroid=[]
    for c in set(new_df1['cluster']):     #return unique values of the clusters
        current_cluster=new_df1[new_df1['cluster']==c][new_df1.columns[:-1]]
        current_mean = current_cluster.mean(axis=0)
        new_centroid.append(current_mean)

    return new_centroid
```

```python
import cv2
import matplotlib.image as mpimg
im_original = cv2.imread('re_resize.jpg')
```

```python
im_original.shape
```

```python
im_updated= (im_original/255).reshape(100*80,3)
```

```python
#no of iteration
n = 15
```

```python
for k in range(1,26):
    init_centroids = random.sample(range(0,len(im_original)),k)
    centroids=[]
    for i in init_centroids:
        centroids.append(im_updated[i])
    im = np.array(im_updated)
    centroid_iteration=[]
    print("K= ",str(k),": ",end=' ')
    for i in range(n):
        print(i,end = ' ')
        get_centroid= findclosetcentroids(centroids,im)  #we have obtained the cluster number for each obervation
        centroids=calc_centroids(get_centroid,im)
        if(i==0):
            centroid_iteration.append(centroids)
        if(i==n//2):
            centroid_iteration.append(centroids)
    centroid_iteration.append(centroids)    
    print()
    im_recovered_0 = im.copy()
    for i in range(len(im)):
        im_recovered_0[i] = centroid_iteration[0][get_centroid[i]]
    im_recovered_0 = im_recovered_0.reshape(100,80,3)

    im_recovered_1 = im.copy()
    for i in range(len(im)):
        im_recovered_1[i] = centroid_iteration[1][get_centroid[i]]
    im_recovered_1 = im_recovered_1.reshape(100,80,3)

    im_recovered_2 = im.copy()
    for i in range(len(im)):
        im_recovered_2[i] = centroid_iteration[2][get_centroid[i]]
    im_recovered_2 = im_recovered_2.reshape(100,80,3)

    fig,ax = plt.subplots(1,4)
    plt.title('For k='+str(k))
    ax[0].imshow(im_original)
    ax[1].imshow(im_recovered_0)
    ax[2].imshow(im_recovered_1)
    ax[3].imshow(im_recovered_2)
```
